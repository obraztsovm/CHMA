Вот ответы на все вопросы к защите:

---

**1. Типы данных**

Данные бывают количественные (числовые измерения — рост, температура) и качественные (категории — пол, цвет). Количественные делятся на дискретные (целые значения, как число событий у Пуассона) и непрерывные (любые значения в интервале, как нормальное или равномерное распределение).

---

**2. Методы создания выборки**

Простая случайная выборка — каждый элемент генеральной совокупности имеет равную вероятность попасть в выборку. В нашей работе использовался метод Монте-Карло: генерация псевдослучайных чисел через `numpy.random` с фиксированным seed для воспроизводимости.

---

**3. Алгоритм группировки данных**

1. Найти размах: R = x_max − x_min
2. Определить число интервалов k по формуле Стёрджеса: k = 1 + log₂(n)
3. Вычислить ширину интервала: h = R / k
4. Построить границы интервалов и подсчитать частоты попадания

В нашей работе при n=100 получилось k=8 интервалов.

---

**4. Графическое представление выборочных данных**

В работе построены три типа графиков. Полигон частот — ломаная линия, соединяющая середины интервалов с соответствующими частотами. Ящик с усами — показывает медиану, квартили Q1 и Q3, выбросы. Эмпирическая функция распределения — ступенчатый график F*(x), показывающий долю наблюдений, не превышающих x.

---

**5. Выборочные числовые характеристики**

- **Начальные моменты** νₖ = (1/n) Σ xᵢᵏ
- **Центральные моменты** μₖ = (1/n) Σ (xᵢ − x̄)ᵏ
- **Среднее** x̄ = ν₁ — центр распределения
- **Медиана** — значение, делящее выборку пополам
- **Мода** — наиболее часто встречающееся значение
- **Исправленная дисперсия** S² = Σ(xᵢ − x̄)² / (n−1) — несмещённая оценка
- **Асимметрия** γ₁ = μ₃ / μ₂^(3/2) — характеризует скошенность
- **Эксцесс** γ₂ = μ₄ / μ₂² − 3 — характеризует остроту пика

---

**6. Эмпирическая функция распределения**

F*(x) = (число наблюдений < x) / n

Это ступенчатая функция, которая в точке каждого наблюдения делает скачок на 1/n. Является несмещённой и состоятельной оценкой истинной функции распределения F(x).

---

**7. Методы построения точечных оценок**

Два основных метода. Метод моментов — приравниваем выборочные моменты к теоретическим и решаем уравнения (например, x̄ = λ для Пуассона). Метод максимального правдоподобия (ММП) — находим параметры, максимизирующие функцию правдоподобия L(θ). В нашей работе оба метода дают одинаковые оценки для всех трёх распределений.

---

**8. Свойства точечных оценок**

- **Несмещённость**: M[θ̂] = θ — оценка не имеет систематической ошибки (поэтому используем S² с делением на n−1, а не на n)
- **Состоятельность**: θ̂ → θ при n → ∞ — с ростом выборки оценка сходится к истинному значению
- **Эффективность**: дисперсия оценки минимальна среди всех несмещённых оценок (достигает границы Крамера-Рао)

---

**9. Интервальные оценки**

Доверительный интервал — это интервал (θ₁, θ₂), который с вероятностью 1−α накрывает истинное значение параметра. В работе построены:
- для μ нормального: через t-распределение Стьюдента с n−1 степенями свободы
- для σ²: через распределение χ²
- для λ экспоненциального: точный интервал через χ²
- уровень доверия во всех случаях 95% (α = 0.05)

---

**10. Проверка статистических гипотез (основные понятия)**

- **H0** — нулевая гипотеза (то, что проверяем, например λ = 3)
- **H1** — альтернативная гипотеза (λ ≠ 3)
- **Уровень значимости α** — вероятность отвергнуть верную H0 (ошибка I рода), в работе α = 0.05
- **Тестовая статистика** — функция от выборки с известным распределением при H0
- **p-value** — вероятность получить такое же или более экстремальное значение статистики при верной H0. Если p-value < α → H0 отвергается
- **Критическая область** — значения статистики, при которых H0 отвергается

В работе использовался двусторонний Z-критерий: H0 отвергается если |Z| > 1.96.

---

**11. Разыгрывание дискретной случайной величины и метод Монте-Карло**

Метод Монте-Карло — численный метод, основанный на генерации случайных чисел. Алгоритм разыгрывания дискретной СВ: генерируем равномерное U ~ U(0,1), затем находим такое значение xₖ, что F(xₖ₋₁) < U ≤ F(xₖ) — это и есть реализация СВ. В работе этот принцип реализован через `numpy.random.poisson`, `numpy.random.normal` и т.д., которые внутри используют именно этот подход.
